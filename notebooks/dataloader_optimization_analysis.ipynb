```python
PATH = './dataset/TUSimple'

class LaneDataset(torch.utils.data.Dataset):
    def __init__(self, dataset_path= PATH, train=True, size=(800, 360)):
        self._dataset_path = dataset_path
        self._mode = "train" if train else "test"
        self._image_size = size # w, h
        self._data = []

        if self._mode == "train":
            file_path = "train_val_gt.txt"
        elif self._mode == "test":
            file_path = "test_gt.txt"
        self._process_list(os.path.join(self._dataset_path, "train_set/seg_label/list", file_path))
            
    def __getitem__(self, idx):
        img_path = self._dataset_path + ("/train_set" if self._mode == "train" else "/test_set") + self._data[idx][0]
        image = cv2.imread(img_path)
        h, w, c = image.shape
        raw_image = image
        image = cv2.resize(image, self._image_size, interpolation=cv2.INTER_LINEAR)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        
        ins_segmentation_path = self._dataset_path + "/train_set" + self._data[idx][1]
        ins_segmentation_image = cv2.imread(ins_segmentation_path)
        ins_segmentation_image = ins_segmentation_image[:, :, 0]
        ins_segmentation_image = cv2.resize(ins_segmentation_image, self._image_size, interpolation=cv2.INTER_LINEAR)
        
        segmentation_image = ins_segmentation_image.copy()
        segmentation_image[segmentation_image > 0] = 1
        
        image = torch.from_numpy(image).float().permute((2, 0, 1))
        segmentation_image = torch.from_numpy(segmentation_image.copy()).to(torch.int64)
        ins_segmentation_image =  torch.from_numpy(ins_segmentation_image.copy())
        
        exists = [int(i) for i in self._data[idx][2]]
        exists = torch.as_tensor(exists)
        
        output = {
            'img_path' : img_path,
            'img' : image,
            "meta" : { "full_img_path" : img_path ,
                     "img_name" :  self._data[ idx ][ 0 ]},
            'segLabel' : segmentation_image,
            'IsegLabel' : ins_segmentation_image,
            'exist' : exists,
            "original_image" : raw_image,
            "label" : segmentation_image
        }
        
        return output
    
    def probmap2lane(self, seg_pred, exist, resize_shape=(720, 1280), smooth=True, y_px_gap=10, pts=56, thresh=0.6):
        """
        Arguments:
        ----------
        seg_pred:      np.array size (5, h, w)
        resize_shape:  reshape size target, (H, W)
        exist:       list of existence, e.g. [0, 1, 1, 0]
        smooth:      whether to smooth the probability or not
        y_px_gap:    y pixel gap for sampling
        pts:     how many points for one lane
        thresh:  probability threshold
    
        Return:
        ----------
        coordinates: [x, y] list of lanes, e.g.: [ [[9, 569], [50, 549]] ,[[630, 569], [647, 549]] ]
        """
        if resize_shape is None:
            resize_shape = seg_pred.shape[1:]  # seg_pred (5, h, w)
        _, h, w = seg_pred.shape
        H, W = resize_shape
        coordinates = []
    
        for i in range(self.cfg.num_classes - 1):
            prob_map = seg_pred[i + 1]
            if smooth:
                prob_map = cv2.blur(prob_map, (9, 9), borderType=cv2.BORDER_REPLICATE)
            coords = self.get_lane(prob_map, y_px_gap, pts, thresh, resize_shape)
            if self.is_short(coords):
                continue
            coordinates.append(
                [[coords[j], H - 10 - j * y_px_gap] if coords[j] > 0 else [-1, H - 10 - j * y_px_gap] for j in
                 range(pts)])
    
    
        if len(coordinates) == 0:
            coords = np.zeros(pts)
            coordinates.append(
                [[coords[j], H - 10 - j * y_px_gap] if coords[j] > 0 else [-1, H - 10 - j * y_px_gap] for j in
                 range(pts)])
        #print(coordinates)
    
        return coordinates
    
    def _process_list(self, file_path):
        with open(file_path) as f:
            for line in f:
                words = line.split()
                image = words[0]
                segmentation = words[1]
                exists = words[2:]
                self._data.append((image, segmentation, exists)) 
                
    def _show_sample_dataset( self, number_samples = 10 ):
        
        # Visualizing some Lane Detection dataset
    
        sns.set_theme()

        f, axarr = plt.subplots( number_samples   ,3 , figsize = ( 20 , 30 ))
        
        plt.axis('off')

        for i in range( number_samples ):

            axarr[ i , 0].imshow(  self.__getitem__( idx = i )[ "original_image" ]  )
            axarr[ i , 0 ].set_title( "Lane Image Data No " + str( i + 1) )
            axarr[ i , 0 ].set_axis_off()
            
            axarr[ i , 1 ].imshow(  self.__getitem__( idx = i )[ "segLabel" ] )
            axarr[ i , 1 ].set_title( "Lane Image Segmentation Data No " + str( i + 1) )
            axarr[ i , 1 ].set_axis_off()
            
            axarr[ i , 2 ].imshow(  self.__getitem__( idx = i )[ "IsegLabel" ] )
            axarr[ i , 2 ].set_title( "Lane Image Segmentation Data No " + str( i + 1) )
            axarr[ i , 2 ].set_axis_off()

        f.tight_layout()
        plt.show()
                
    def __len__(self):
        return len(self._data)  
```


This code tests the performance of PyTorch's `DataLoader` with different configurations of `batch_size` and `num_workers`. The objective is to measure the time taken to load an entire dataset for various settings and to visualize the results using a heatmap.

---

#### **Code Breakdown:**

- **Setup for Testing (`output_dir` and `test_dataloader` function):**
  - A directory `Dataloader_Results` is created to store output results.
  - The `test_dataloader` function measures the time required to load a dataset using a `DataLoader` with given `batch_size` and `num_workers`.

- **Configurations to Test (`batch_sizes` and `worker_configs`):**
  - Different values of `batch_size` (8, 16, 32, 64) and `num_workers` (8, 4, 2, 0) are set to test various configurations.

- **Testing Loop:**
  - Nested loops iterate through combinations of `batch_size` and `num_workers`.
  - Time taken for each configuration is recorded, and exceptions are handled to ensure robust testing.

- **Visualization:**
  - Results are stored and structured into a matrix format to prepare for a heatmap.
  - A heatmap is generated using Seaborn to visually analyze the relationship between `batch_size`, `num_workers`, and data loading time.
  - The heatmap is saved as a PNG file in the `Dataloader_Results` directory.

---

#### **Key Visual Output:**
- **Heatmap:**
  - Displays the performance of different combinations of `num_workers` and `batch_size`.
  - Helps in identifying the optimal configuration for faster data loading.

```python
import os
import time
import matplotlib.pyplot as plt
import seaborn as sns
from torch.utils.data import DataLoader

# Ensure output directory
output_dir = "Dataloader_Results"
os.makedirs(output_dir, exist_ok=True)

# Test configurations
batch_sizes = [8, 16, 32, 64]
worker_configs = [8, 4, 2, 0]
results = []

# Test function
def test_dataloader(batch_size, num_workers):
    dataset = LaneDataset(dataset_path="./dataset/TUSimple", train=True, size=(800, 360))
    dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True, num_workers=num_workers)

    start_time = time.time()
    try:
        for i, batch in enumerate(dataloader):
            pass  # Simulate full dataset loading
    except Exception as e:
        print(f"Error during DataLoader test: {e}")
        return float("inf")  # Mark failed test as infinity time
    return time.time() - start_time

# Run tests
for workers in worker_configs:
    for batch_size in batch_sizes:
        print(f"Testing: Workers={workers}, Batch Size={batch_size}")
        time_taken = test_dataloader(batch_size=batch_size, num_workers=workers)
        print(f"Completed in {time_taken:.2f} seconds")
        results.append((workers, batch_size, time_taken))

# Generate heatmap
heatmap_data = {}
for workers, batch_size, time_taken in results:
    if workers not in heatmap_data:
        heatmap_data[workers] = {}
    heatmap_data[workers][batch_size] = time_taken

worker_keys = sorted(heatmap_data.keys())
batch_keys = sorted(batch_sizes)

heatmap_matrix = [[heatmap_data[w][b] for b in batch_keys] for w in worker_keys]

sns.heatmap(
    heatmap_matrix,
    annot=True,
    fmt=".2f",
    xticklabels=batch_keys,
    yticklabels=worker_keys,
    cmap="viridis"
)
plt.title("DataLoader Performance: Workers vs Batch Size")
plt.xlabel("Batch Size")
plt.ylabel("Workers")
plt.tight_layout()
heatmap_path = os.path.join(output_dir, "dataloader_performance_heatmap.png")
plt.savefig(heatmap_path)
print(f"Saved visualization: {heatmap_path}")
plt.show()
```

### **Conclusions from Heatmap**

1. **Effect of Number of Workers:**
   - Increasing the number of workers significantly improves performance. 
   - **8 workers** consistently achieve the fastest data loading times (~36 seconds), while **0 workers** are the slowest (154–171 seconds).

2. **Effect of Batch Size:**
   - Larger batch sizes (**32, 64**) result in better performance, especially when paired with higher worker counts.
   - Smaller batch sizes (**8, 16**) take slightly more time due to more frequent data loading operations.

3. **Optimal Configuration:**
   - The best combination is **8 workers** with a batch size of **64**, providing the shortest loading time.

4. **General Recommendations:**
   - Use **8 workers** for optimal data loading efficiency.
   - Pair larger batch sizes (**32, 64**) with higher worker counts for balanced performance. 
   - Avoid **0 workers**, as it is highly inefficient for larger datasets.
